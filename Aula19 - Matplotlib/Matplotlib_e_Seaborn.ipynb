```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
plt.plot([1,3,5],[2,4,7])
plt.show()
```

```python
x = [10,20,30,40,50]
y = [100,400,900,1600,2500]
plt.plot(x,y)
plt.xlabel('Bases')
plt.ylabel('Potências')
plt.title('Potenciação')
plt.show()
```

```python
x2 = [1,3,4,5,6,7]*2
y2 = [2,4,6,7,8,10]* 2
plt.plot(x2, y2, label="Gráfico com Matplotlib")
plt.legend()
plt.show()
```

```python
x = [2,4,6,8,10]
y = [1,3,5,7,9]
```

```python
plt.bar(x, y, label="Barras", color='green')
plt.legend()
plt.show()
```
